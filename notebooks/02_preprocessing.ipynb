# ‚öôÔ∏è Titanic Data Preprocessing

**Objective:** Clean and prepare data for machine learning models.


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/raw/train.csv")
```

## üóëÔ∏è Dropping Irrelevant Features


```python
df.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis=1, inplace=True)
```

## üß© Handling Missing Values


```python
df["Age"].fillna(df["Age"].median(), inplace=True)
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
```

## üî† Encoding Categorical Features


```python
df = pd.get_dummies(df, drop_first=True)
```

## üéØ Feature and Target Separation


```python
X = df.drop("Survived", axis=1)
y = df["Survived"]
```

## üìè Feature Scaling


```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## üíæ Saving Processed Dataset


```python
processed_df = pd.DataFrame(X_scaled, columns=X.columns)
processed_df["Survived"] = y.values

processed_df.to_csv("../data/processed/processed_data.csv", index=False)
```

## ‚úÖ Preprocessing Summary
- Removed unnecessary columns
- Handled missing values
- Encoded categorical variables
- Applied feature scaling
- Data ready for model training
